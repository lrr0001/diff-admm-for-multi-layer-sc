```python
import tensorflow
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.models import Model
```

```python
x = np.reshape(np.array(range(20)),(20,1,1,1))
y = 2*x
#z = np.concatenate((x,y),axis=1)
#print(z)
```

```python
from tensorflow.keras.layers import Layer
class LinearScalar(Layer):

  def __init__(self,**kwargs):
    super(LinearScalar, self).__init__(**kwargs)
  def build(self, input_shape):
    self.kernel = self.add_weight(name='kernel', 
                                      shape=(1, 1, 1, 1),
                                      initializer='uniform',
                                      trainable=True)
    super(LinearScalar, self).build(input_shape)  # Be sure to call this at the end

  def call(self, inputs):
    b = tensorflow.constant(value=1.0,dtype=tensorflow.float32,shape=(1,1,1,1))
    y = self.kernel*inputs
    z = b*inputs
    #tensorflow.losses.add_loss(tensorflow.nn.l2_loss(y - b))
    return (y,z-y)
```

```python
inputLayer = tensorflow.keras.Input(shape=(1,1,1))
xta,myerr = LinearScalar()(inputLayer)
model = Model(inputs=inputLayer,outputs=xta)
```

```python
model.add_loss(tensorflow.nn.l2_loss(1.41421356237*myerr))
model.compile(optimizer='rmsprop',loss='mse')
model.fit(x,y,epochs=200,batch_size=1)
```
